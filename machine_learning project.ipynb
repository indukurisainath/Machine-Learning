```python
gender_dict ={'Male':1, 'Female':2}
# mapping values in column from dictionary
df['gender_encod']= df.gender.map(gender_dict)
df = df.drop(['gender'], axis = 1)

restaurants_dict={'Never':0,'1-2 times':1,'3-5 times':2,'5-8 times':3,'8-12 times':4,'more than 12 times':5}
df['restaurants_encod']= df.restaurants.map(restaurants_dict)
df = df.drop(['restaurants'], axis = 1)


malls_dict={'Never':0,'1-2 times':1,'3-5 times':2,'5-8 times':3,'8-12 times':4,'more than 12 times':5,'everyday':6}
df['malls_encod']= df.malls.map(malls_dict)
df = df.drop(['malls'], axis = 1)

outfits_dict={'No':0,'Yes':1}
df['outfits_encod']= df.outfits.map(outfits_dict)
df = df.drop(['outfits'], axis = 1)

amazon_dict={'wont spend':0,'<500':1,'500-1000':2,'1k-3k':3,'3k-5k':4,'more than 5k':5}
df['amazon_encod']= df.amazon.map(amazon_dict)
df = df.drop(['amazon'], axis = 1)

skip_d_dict={'Never':0,'1-2 times':1,'3-5 times':2,'5-10 times':3,'10-15 times':4,'more than 15 times':5,'Daily':6}
df['skip_d_encod']= df.skip_d.map(skip_d_dict)
df = df.drop(['skip_d'], axis = 1)

juices_dict={'Never':0,'once in a month':1,'2-4 times in a month':2,'5-7 times in a month':3,'10-15 times':4,'more than 15 times':5,'Daily':6}
df['juices_encod']= df.juices.map(juices_dict)
df = df.drop(['juices'], axis = 1)

party_dict={'No':0,'Yes':1}
df['party_encod']= df.party.map(party_dict)
df = df.drop(['party'], axis = 1)

cooldrinks_dict={'wont drink':0,'1-2 times':1,'3-5 times':2,'5-10 times':3,'10-20 times':4,'more than 20 times':5,'Daily':6}
df['cooldrinks_encod']= df.cooldrinks.map(cooldrinks_dict)
df = df.drop(['cooldrinks'], axis = 1)

snacks_dict={'Never':0,'1-2 times':1,'3-5 times':2,'5-8 times':3,'8-12 times':4,'12-15 times':5,'15-20 times':6,'more than 20 times':7,'Daily':8}
df['snacks_encod']= df.snacks.map(snacks_dict)
df = df.drop(['snacks'], axis = 1)

df
```

```python
#Let's split our data
x = df.drop(['Otpt'], axis=1) #Input features
y= df["Otpt"] #Output
```

```python
# Split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=40)
```
